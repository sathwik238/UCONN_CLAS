# Data Clustering and Predictive Modeling Project

## Objective
This project aims to:
1. Cluster data into 3 or 4 groups based on covariates and analyze their relationships with the outcome variable \( y \).
2. Develop a predictive function \( \hat{y} \) to approximate \( y \) more effectively, as \( y \) is poorly described.

---

## Dataset Overview
- **Outcome Variable (y):** The target variable; \( y = 0 \) rows are filtered out.
- **Covariates:**
  - \( X1, X2, X3 \): Measure similar aspects.
  - \( Z1, Z2 \): Measure another aspect but are closely related.
  - \( A \): Represents a distinct attribute.
  - \( B \): Represents another distinct attribute.

## EDA

```python
for i in data.columns:
    if i == 'ID':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(data[i])
    plt.title(i)
    plt.show()
```

```python
#scatter plot to check for relation between the variables and outcome variable
for i in data.columns:
    if i == 'ID':
        continue
    plt.figure(figsize=(8,4))
    sns.scatterplot([data[i],data['Y']])
    plt.title(i)
    plt.show()
```

```python
data.describe()
```

```python
# correlation matrix
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

for i in data.columns:
    if i == 'ID' or i == 'Y':  # Skip ID and Y itself
        continue
    try:
        # Ensure at least 5 unique values for meaningful binning
        if data[i].nunique() < 5:
            continue  

        # Use quantile-based binning with unique edges
        data['bins'] = pd.qcut(data[i], q=5, labels=False, duplicates='drop')

        # Plot the average outcome variable per bin
        data.groupby('bins')['Y'].mean().plot(marker='o', linestyle='-')  
        plt.xlabel("bins")
        plt.ylabel("Average Y")
        plt.title(i)
        plt.show()
        
    except Exception as e:
        print(f"Skipping {i} due to error: {e}")

```
